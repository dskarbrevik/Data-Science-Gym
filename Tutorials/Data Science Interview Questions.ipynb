**Fizzbuzz+**

fizzbuzz is a simple for loop + if/else condition problem, but we'll add an additional twist.

```python
def fizzbuzzbasic(n):
    for i in range(n):
        if ((i+1) % 3 == 0) and ((i+1) % 5 == 0):
            print((i+1), "Fizzbuzz!")
        elif (i+1) % 3 == 0:
            print((i+1), "Fizz")
        elif (i+1) % 5 == 0:
            print((i+1), "Buzz")
        else:
            print(i+1)
```

```python
fizzbuzzbasic(20)
```

```python
def fizzbuzzplus(n:int, word_numbers:list):
    
    for i in range(n):
        check = False
        for j in range(len(word_numbers)):
            if (i+1) % word_numbers[j][0] == 0:
                print((i+1),word_numbers[j][1])
                check = True
        if not check:
            print(i+1)
```

```python
fizzbuzzlist = [(3,"fizz"),(5,"buzz"),(7,"fizzbuzz")]

fizzbuzzplus(20, fizzbuzzlist)
```

**t distribution**

```python
student1 = np.random.standard_t(df=3, size=100000)
student2 = np.random.standard_t(df=30, size=100000)

print("Distribution = {}".format("t distribution"))
print("-"*30)
print("When df = 3")
print("Mean = {:.2f}".format(np.abs(np.mean(student1))))
print("SD = {:.2f}".format(np.std(student1)))
print("")
print("When df = 30")
print("Mean = {:.2f}".format(np.abs(np.mean(student2))))
print("SD = {:.2f}".format(np.std(student2)))

fig, ax = plt.subplots(figsize=(16,8))
plt.xlim(-5,5)

sns.distplot(student1, hist=False, label="df = 3", ax=ax)
sns.distplot(student2, hist=False, label="df = 30", ax=ax)
ax.set_title("t distribution", fontsize=25);

plt.setp(ax.get_legend().get_texts(), fontsize='20');
```

The t distribution changes with changing degrees of freedom. As df becomes higher, the t distribution essentially becomes the normal distribution. Notice how the variance is higher when degrees of freedom is lower.

**Poisson distribution**

```python
poisson1 = np.random.poisson(lam=2,size=100000)
poisson2 = np.random.poisson(lam=5,size=100000)
poisson3 = np.random.poisson(lam=10,size=100000)
poisson4 = np.random.poisson(lam=20,size=100000)

print("Distribution = {}".format("Poisson"))
print("-"*30)
print("When mean = 2")
print("SD = {:.2f}".format(np.std(poisson1)))
print("")
print("When mean = 5")
print("SD = {:.2f}".format(np.std(poisson2)))
print("")
print("When mean = 10")
print("SD = {:.2f}".format(np.std(poisson3)))
print("")
print("When mean = 20")
print("SD = {:.2f}".format(np.std(poisson4)))


fig, ax = plt.subplots(figsize=(16,8))
sns.distplot(poisson1, kde=False, label="$\mu$ = 2", ax=ax)
sns.distplot(poisson2, kde=False, label="$\mu$ = 5", ax=ax)
sns.distplot(poisson3, kde=False, label="$\mu$ = 10", ax=ax)
sns.distplot(poisson4, kde=False, label="$\mu$ = 20", ax=ax)
ax.set_title("Poisson distribution", fontsize=25);
plt.legend();
plt.xlim(0,30);
plt.setp(ax.get_legend().get_texts(), fontsize='20');
```
