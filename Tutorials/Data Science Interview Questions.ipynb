**Fizzbuzz+**

fizzbuzz is a simple for loop + if/else condition problem, but we'll add an additional twist.

```python
def fizzbuzzbasic(n):
    for i in range(n):
        if ((i+1) % 3 == 0) and ((i+1) % 5 == 0):
            print((i+1), "Fizzbuzz!")
        elif (i+1) % 3 == 0:
            print((i+1), "Fizz")
        elif (i+1) % 5 == 0:
            print((i+1), "Buzz")
        else:
            print(i+1)
```

```python
fizzbuzzbasic(20)
```

```python
def fizzbuzzplus(n:int, word_numbers:list):
    
    for i in range(n):
        check = False
        for j in range(len(word_numbers)):
            if (i+1) % word_numbers[j][0] == 0:
                print((i+1),word_numbers[j][1])
                check = True
        if not check:
            print(i+1)
```

```python
fizzbuzzlist = [(3,"fizz"),(5,"buzz"),(7,"fizzbuzz")]

fizzbuzzplus(20, fizzbuzzlist)
```

**Create and plot data for the following probability distributions (also write a sentence or two about the distribution):**

* Uniform
* Normal
* t
* Poisson 
* Exponential

```python
%matplotlib inline

import numpy as np
import pandas as pd

from scipy import stats, integrate

import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)

import warnings
warnings.filterwarnings("ignore")
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(16,16));

uniform = np.random.rand(100000)
ax[0][0] = sns.distplot(uniform, kde=False, ax=ax[0][0])
ax[0][0].set(xlim=(0,1), ylim=(0,4000))
ax[0][0].set_title("uniform distribution", fontsize=20)


normal = np.random.randn(100000)
ax[0][1] = sns.distplot(normal, bins=np.linspace(-4,4,30), ax=ax[0][1])
ax[0][1].set_title("normal distribution", fontsize=20)


student1 = np.random.standard_t(df=3, size=100000)
student2 = np.random.standard_t(df=30, size=100000)
ax[1][0] = sns.distplot(student1, hist=False, label="df = 3", ax=ax[1][0])
ax[1][0] = sns.distplot(student2, hist=False, label="df = 30", ax=ax[1][0])
ax[1][0].set(xlim=(-5,5), title="t")
ax[1][0].set_title("t distribution", fontsize=20)
plt.setp(ax[1][0].get_legend().get_texts(), fontsize=15)


exponential = np.random.exponential(size=100000)
ax[1][1] = sns.distplot(exponential, kde=False, ax=ax[1][1])
ax[1][1].set(xlim=(0,8))
ax[1][1].set_title("exponential distribution", fontsize=20)

fig,ax = plt.subplots(figsize=(16,8))
poisson1 = np.random.poisson(lam=2,size=100000)
poisson2 = np.random.poisson(lam=5,size=100000)
poisson3 = np.random.poisson(lam=10,size=100000)
poisson4 = np.random.poisson(lam=20,size=100000)

sns.distplot(poisson1, kde=False, label="$\mu$ = 2", ax=ax)
sns.distplot(poisson2, kde=False, label="$\mu$ = 5", ax=ax)
sns.distplot(poisson3, kde=False, label="$\mu$ = 10", ax=ax)
sns.distplot(poisson4, kde=False, label="$\mu$ = 20", ax=ax)
plt.legend();
ax.set(xlim=(0,30));
ax.set_title("poisson distribution", fontsize=20);
plt.setp(ax.get_legend().get_texts(), fontsize=15);
```

**Uniform:** Possibly the simplist distribution, all possible outcomes occur with equal likelihood. For example, the roll of a die.

**Normal:** Most common distribution in nature.

**t:** The t distribution changes with changing degrees of freedom. As df becomes higher, the t distribution essentially becomes the normal distribution. Notice how the variance is higher when degrees of freedom is lower.

**Poisson:** Looks like a normal distribution right? But notice how all values are descrete.

**exponential:** As the name implies, this distribution tails off exponentially.
